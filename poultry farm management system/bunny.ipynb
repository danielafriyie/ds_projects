### Sales / Purchase
    - Name
    - Phone
    - Address
    - Category
    - Date
    - Invoice No.
    - Quantity
    - Unit Price
    - Amount
    - Description

```python
def random_date():
    year = 2019
    month = np.random.randint(1, 12)
    day = np.random.randint(1, 30) if month != 2 else np.random.randint(1, 28)
    date = f'{day}-{month}-{year}'
    return date

def random_time():
    time = ['8:30', '12:30', '16:30']
    return ''.join(random.sample(time, 1))
```

```python
def sales_purchase_data():
    data = []
    
    suppliers = [
        'Daniel Afriyie', 'David Sarpong', 'Mamittah Opoku Mensah',
        'Leah Opii Birago', 'Anthony Mensah', 'Davis Ampah',
        'Yaw Mensah', 'Akwasi Boo', 'Yaa Attaa', 
        'Kwasi Aidoo'
    ]
    suppliers_dict = {}
    for ss in suppliers:
        suppliers_dict[ss] = ''.join([str(x) for x in np.random.randint(0, 9, 10).tolist()])
    for name, phone in suppliers_dict.items():
        for _ in range(np.random.randint(25, 70)):
            transaction = {}
            qty = np.random.randint(250, 1000)
            up = ''.join([str(x) for x in np.round(np.random.random(1) * np.random.randint(10, 500), 2)])
            amt = np.round(float(up) * qty, 2)
            transaction['Name'] = name
            transaction['Phone'] = phone
            transaction['Address'] = ''
            transaction['Category'] = 'Purchase'
            transaction['Date'] = random_date()
            transaction['Invoice No.'] = f"{name[:3].upper()}{''.join([str(x) for x in np.random.randint(0, 9, 10).tolist()])}"
            transaction['Quantity'] = qty
            transaction['Unit Price'] = up
            transaction['Amount'] = amt
            transaction['Description'] = f"{qty} purchases"
            data.append(transaction)
            
    debtors = [
        'Afia Agyeiwaa', 'Kojo Manu', 'Evans Mensah',
        'Yaw Kwaah', 'Andy Koo', 'Tutu Gyau',
        'Kofi Koo', 'Kwaku Appiah', 'Kobby Bi',
        'Kojo Afrifa', 'Arthur Mensah', 'Agyemang Stephen',
        'Esther Agyemang', 'Felix Manu', 'Opoku Francis',
        'Kwame Adu', 'Yaw Baah', 'Iddrissu Abugri', 
        'Felicia Joll'
    ]
    debtors_dict = {}
    for ds in debtors:
        debtors_dict[ds] = ''.join([str(x) for x in np.random.randint(0, 9, 10).tolist()])
    debtors_data = []
    for name, phone in debtors_dict.items():
        for _ in range(np.random.randint(25, 70)):
            transaction = {}
            qty = np.random.randint(250, 1000)
            up = ''.join([str(x) for x in np.round(np.random.random(1) * np.random.randint(10, 500), 2)])
            amt = np.round(float(up) * qty, 2)
            transaction['Name'] = name
            transaction['Phone'] = phone
            transaction['Address'] = ''
            transaction['Category'] = 'Sale'
            transaction['Date'] = random_date()
            transaction['Invoice No.'] = f"PFMS{''.join([str(x) for x in np.random.randint(0, 9, 10).tolist()])}"
            transaction['Quantity'] = qty
            transaction['Unit Price'] = up
            transaction['Amount'] = amt
            transaction['Description'] = f"{qty} sales"
            data.append(transaction)
    return data
```

```python
def birds_data():
    data = []
    cols = ['Pen House', 'Date', 'Time', 'Quantity']
    for _ in range(1200):
        collection = {}
        collection['Pen House'] = np.random.randint(1, 8)
        collection['Date'] = random_date()
        collection['Time'] = random_time()
        collection['Quantity'] = np.random.randint(150, 500)
        data.append(collection)
    return data
```
